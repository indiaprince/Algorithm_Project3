<h2>You are given a sorted array S[1, … , n] with n distinct integers, i.e., S[i] < S[i + 1], for all 1 ≤ i < n.
Design a divide-and-conquer algorithm to decide whether there exists an index k such that S[k] = k.</h2>

## - Propose an Approach

<h2>Characterisitcs of the Array</h2>


- <h3>Sorted</h3>
- <h3>Every elements are distnct.</h3>
- <h3> S[i] < S[i+1], for all 1 <= i < b ▶▶ it is increasing sequence</h3>

```python
import random

S_unsorted = random.sample(range(1, 16), 10)
print(S_unsorted)
S  = sorted(S_unsorted)
N=len(S)
print(S)
```

<h2>To solve this problem, we can think of the following algorithms</h2>

<h3>Bruteforce Search</h3>
- Worst Case : O(N)

```python
def Bruteforce_Search(S):
  for k in range(1, N):
    if(S[k-1]==k): return True
  return False

if(Bruteforce_Search(S)): print(f'Array S has an index k such taht S[k]=k')
else : print(f'Array S does not have an index k such taht S[k]=k')
```

<h3>Divide&Conquer(Since it is sorted)</h3>
-Worst Case : O(logN)

IDEA : 

for some k, if S[k] < k, as it is increasing sequence, it is not possible to get S[k] == k. Because element in S will not be decreased.

if S[k]==k, it returns true

However if S[k] > k, there are some possibilities that S[k] == k, so we can divide the problem.

Q : 그냥 첫 index 확인하고 끝내면 되지 않을까?? 첫번째 인덱스 다르면 같을 일 없는 것 같은데

```python
def Binary_Search(S,k):
  if(k<1) : return False
  if(S[k-1]==k) : return True
  elif(S[k-1]<k): return False
  else : return Binary_Search(S,k//2)
Binary_Search(S, len(S)//2)
```
